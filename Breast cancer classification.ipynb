```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
import warnings
warnings.filterwarnings("ignore")
```

```python
breast_cancer_data = load_breast_cancer()
breast_cancer_data
```

```python
breast_cancer_data.keys()
```

```python
data = pd.DataFrame(breast_cancer_data.data,columns=breast_cancer_data.feature_names)
data.head()
```

```python
data["labels"] = breast_cancer_data.target
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
X = data.drop(["labels"],axis=1)
y = data["labels"]
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=.1,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_std = scaler.fit_transform(x_train)
X_test_std = scaler.transform(x_test)
```

```python
#importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#setting up the layers of neural networks

model = keras.Sequential([
                        keras.layers.Flatten(input_shape=(30,)),
                        keras.layers.Dense(20,activation="relu"),
                        keras.layers.Dense(2,activation="sigmoid")
])
```

```python
#compiling the neural network

model.compile(optimizer="adam",
             loss="sparse_categorical_crossentropy",
             metrics=["accuracy"])
```

```python
history = model.fit(X_train_std,y_train,validation_split=.1,epochs=50)
```

```python
#visualizing the accuracy and loss

plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.title("Model accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(["training data","val_data"],loc="lower right")
plt.show()
```

```python
#visualizing the accuracy and loss

plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("Model loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend(["training data","val_data"],loc="upper right")
plt.show()
```

```python
#Accuracy of the model on test data
loss,accuracy = model.evaluate(X_test_std,y_test)
loss,accuracy
```

```python
#Building the predictive system

input_df = pd.DataFrame(x_test.iloc[0])
input_data_array = np.asarray(input_df).reshape(1,-1)
input_data = scaler.transform(input_data_array)
prediction = model.predict(input_data)
prediction_label = [np.argmax(prediction)]

print(prediction)
print(prediction_label)

if prediction_label[0] == 0:
    print("Tumor is maligant.")
else:
    print("Tumor is benign.")
```
